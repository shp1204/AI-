# 1. 모듈러 연산

정수론과 소수
정수론(Number theory)은 이름에 걸맞게 수 그 자체에 대해 탐구하는 수학의 한 분야입니다.
수학자 가우스가 “수학은 과학의 여왕이요, 정수론은 수학의 여왕이다.”라고 말했을 정도로 정수론은 수학에 있어 중요한 학문이며 수학과 관련된 거의 모든 분야에 깊이 관여합니다. 이는 컴퓨터과학에 있어서도 예외는 아닙니다. 본래 컴퓨터는 “수”를 계산하기 위해 탄생하였으며 0과 1로 이루어진 디지털 세계에서 모든 정보들은 “수”로 다루어집니다. 그래서 컴퓨터과학에 있어 정수론의 중요성은 아무리 강조해도 지나치지 않습니다.

우리가 인지하지 않았을 뿐 우리는 이미 초등학교, 중학교 때부터 최대 공약수, 피타고라스의 정리 등을 통해 정수론을 배워왔습니다. 그 중에서도 정수론에서 매우 중요하고 흥미롭게 다루는 수가 바로 소수(Prime number)입니다. 소수는 그 불규칙한 특성으로 인해 아직까지도 많은 수학자에 의해 연구되고 있는 분야입니다. 이 장에서는 정수론과 소수에 대해 함께 탐구해봅시다.

모듈러 연산
소수의 정의를 알아보기 전에 먼저 소개할 것이 하나 있습니다. 나머지 연산이라고도 하는 모듈러 연산(Modular arithmetic)입니다. 모듈러 연산은 a mod b = c와 같이 표기 하며 뜻은 “a를 b로 나눈 나머지는 c”가 됩니다. 10 mod 3은 10을 3으로 나눈 나머지 이므로 1이 됩니다. 파이썬에서는 이러한 모듈러 연산을 퍼센트(%)기호로 나타냅니다.

print(10 % 3)

따라서 위의 코드는 1을 출력하게 됩니다.

```python
def mod(a, b):
    return a % b


# 결과 출력을 위한 코드입니다. 자유롭게 값을 바꿔보며 확인해보세요.
print(mod(10,3))
```

# 2. 소수

소수의 정의

**소수**

소수란 “자신보다 작은 두 개의 자연수를 곱하여 만들 수 없는 1보다 큰 자연수”입니다.

예를 몇 개 들어봅시다.

6: 6=2×3로 나타낼 수 있으므로 소수가 아닙니다.

5: 5=1×5외에는 나타낼 수 없으므로 소수 입니다.

4: 4=2×2로 나타낼 수 있으므로 소수가 아닙니다.

3: 3=1×3외에는 나타낼 수 없으므로 소수 입니다.

2: 2=1×2외에는 나타낼 수 없으므로 소수 입니다.

1: 자기 자신보다 작은 자연수가 없으므로 소수가 아닙니다.

1은 소수가 아니라는 점에 유의해야 합니다.



**합성수**

1보다 큰 자연수 중 소수가 아닌 수를 합성수(Composite number)라고 부릅니다. 1은 소수도 합성수도 아닙니다.

**소수 판별법**

위의 예를 따라가다보면 알 수 있겠지만 소수인지 아닌지를 알아보기 위해서는 자신보다 작은 자연수의 곱이 하나라도 존재하는 지를 확인하면 됩니다. 좀 더 자세한 과정을 살펴봅시다.

p는 소수가 아님을 알고싶다면, p=a×b를 만족하는 p보다 작은 a, b를 찾으면 됩니다. 식을 조금 변형 하면, p÷a=b가 됩니다. 즉 p를 p보다 작은 자연수 a로 나누었을 때 마찬가지로 p보다 작은 자연수 b가 된다면 p는 소수가 아니라는 말이 됩니다.

4를 예를 들어서 확인해 봅시다.

p	a (a < p)	b (=p ÷ a)

4	1	4 (p보다 작지 않음)

4	2	2

4	3	1.3333333…

a가 2일때 b가 2가 되는 것을 확인 할 수 있습니다. 식으로 나타 4÷2=2이므로 4는 소수가 아닙니다.

한편 a가 1일 때를 유의깊게 관찰해 봅시다. a가 1인 경우에는 b는 항상 p와 같으므로 굳이 이 경우를 체크할 필요는 없어보입니다. 따라서 다음 예제에서는 이 경우를 제외하도록 하겠습니다.

이번에는 5를 예로 들어보겠습니다.

p	a (a < p)	b (=p ÷ a)

5	2	2.5

5	3	1.6666666…

5	4	1.25

모든 a에 대해 우리가 원하는 b가 존재하지 않음을 알 수 있습니다. 따라서 5는 소수입니다.

프로그래밍에 적용하기!

이 방법을 프로그래밍으로 구현하기 위해 우리는 모듈러 연산을 유용하게 사용할 수 있습니다.
“p를 a로 나눈 b가 자연수”라는 말은 다른 말로 “p를 a로 나눈 나머지가 0”이라는 말이 되기 때문입니다.
최종적으로 정리해보면, “p>a>1인 자연수 a에 대해 p mod a=0인 a가 존재하면 p는 소수가 아니다”가 됩니다.

```python
def isPrime(n):

    # 1. n이 1인 경우, False를 반환
    if n == 1:
        return False
    
    # 2. n이 합성수이면, False를 반환
    # n을 2부터 n-1 까지 나누었을 때, 어떤 수라도 나눠지면 바로 return
    for i in range(2, n) :
        if n % i == 0 :
            return False
    
    # 3. n이 소수면, True를 반환
    return True
            

# 결과 출력을 위한 코드입니다. 자유롭게 값을 바꿔보며 확인해보세요.
print(isPrime(7))
```

# 3. 에라토스테네스의 체

에라토스테네스의 체
위에서 배운 방법을 통해 어느 한 수가 소수인지 아닌지를 판별할 수 있게 되었습니다.
그런데 만약 “10000보다 작은 모든 소수를 구하라”라는 문제가 있다면 어떻게 해결해야 할까요?
1부터 9999까지 모든 수를 소수 판별 해볼 수도 있지만 이것은 매우 많은 계산을 필요로 할 것 입니다.
다행히 이 같은 문제를 해결하기 위한 효율적인 방법이 이미 있습니다.
에라토스테네스의 체(Sieve of Eratosthenes)라는 알고리즘입니다.

이 방법은 비유하자면 체를 통해 고운 입자만 통과시키는 것처럼 주어진 범위의 수 중에서 소수만을 남기는 방법입니다.
2부터 100까지의 수를 생각해 봅시다. 우리는 2가 소수임을 이미 알고 있습니다. 그렇다면 2로 만든 체를 통해 숫자를 걸러내 봅시다.
숫자 중 2의 배수인 4, 6, 8, 10, 12 …, 100이 체에 의해 걸러집니다. 이 숫자들은 모두 2로 나누었을 때 나머지가 0인 숫자들이므로 소수가 아닙니다.
2 다음인 3 역시 소수입니다. 3으로 만든 체를 통해 남은 숫자들을 걸러내 봅시다. 3, 6, 9, 12 …, 99가 체에 의해 걸러집니다.
이 숫자들 역시 모두 3으로 나누었을때 나머지가 0이므로 소수가 아닙니다.
3 다음인 4는 이미 2에 의해 걸러진 숫자이므로 통과합니다. 5는 아직 걸러지지 않은 숫자이므로 소수임을 알 수 있습니다.



정리해 보면 다음과 같습니다.

찾고자 하는 범위의 자연수를 나열한다.
2부터 시작하여, 2의 배수를 지워나간다.
다음 소수의 배수를 모두 지운다.
이러한 과정을 반복하면 주어진 범위 내의 소수를 모두 찾아낼 수 있습니다.

```python
def eratosthenes(n):
    
    # n개의 True 값이 들어있는 목록을 준비     
    # (True는 소수, False는 합성수를 의미)
    sieve = [True] * n
    
    # 2부터 n까지 하나씩 순차적으로 소수 여부를 판단
    for i in range(2, n):
        
        # 1. i가 소수인 경우 i의 배수를 False로 변경
        for j in range(i+i, n, i) :
            sieve[j] = False
            
        # 2. i는 소수이므로 True값을 유지
        else :
            pass
        
    # 값이 True인 숫자를 추려낸다
    return [i for i in range(2, n) if sieve[i] == True]

# 결과 출력을 위한 코드입니다. 자유롭게 값을 바꿔보며 확인해보세요.
print(eratosthenes(10))
```

# 4. 소인수분해

소인수분해
소인수분해는 어떤 수를 소수들의 곱으로만 나타내는 것을 말합니다. 그 곱셈에 사용되는 소수들을 우리는 소인수라고 부릅니다. 예를 들어 12는 12=2×2×3로 소인수 분해가 됩니다. 소수의 경우에는 자기 자신 외에 나누어지는 소수가 없으므로 그 자체로 소인수분해가 된 것으로 간주합니다. 예를 들어 7은 소수 이므로 7 그 자체로 이미 소인 분해된 것입니다.

프로그래밍에 적용하기!
소인수분해를 프로그래밍으로 구현하는 가장 간단한 방법은 n보다 작은 모든 소수로 나누어 보는 것입니다. 이는 우리가 의무교육 과정에서 배우는 소인수분해와 동일한 방법입니다. 60를 예로 들어 봅시다.

먼저 가장 작은 소수인 2로 나누어 봅니다. 60=2×30

30은 소수가 아니므로 이어서 2로 다시한번 나누어 봅니다. 60=2×2×15

15는 소수가 아닙니다. 더이상 2로는 나눌수 없으므로 다음으로 큰 소수인 3으로 나누어 봅니다. 60=2×2×3×5

5는 소수입니다. 따라서 소인수분해가 완료되었습니다.

이와 같은 과정을 통해 우리는 소인수분해를 할 수 있습니다.
