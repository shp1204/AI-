# 1. 데이터 과학



데이터 과학의 목표 : 의사결정 + 수익화 !

데이터 과학은 현실에서 발생하는 문제를 해결하기 위해서 컴퓨터를 활용하여 데이터를 분석하는 작업 전반을 의미한다

# 2. 머신러닝

데이터 과학의 영역 : 빅데이터 분석, 통계 분석, 머신러닝

* 빅데이터 분석 : 그로스 해킹(Growth Hacking), 프로그래밍을 포함한 기술 분석을 통한 마케팅, 상관관계 분석

* 통계 분석 : 설명, 해석에 중점 

* 머신 러닝 : 예측, 패턴을 분석하는 것, 컴퓨터가 데이터로부터 스스로 학습할 수 있도록 프로그래밍하는 과학적 활동




[ 머신러닝은 언제 사용할까 ? ]
- for hand-tuning or long lists of rules
- for not working and traditional approach
- for adapting to new data


* 데이터 업데이트와 자동화 부분이 다르다

# 4. 머신러닝 개념

## 4_1 개념

[ DATA ]
데이터란 현실 세계의 어떤 현상을 관찰을 통해 기록한 것

[ 데이터에 대한 설명 ]
- 데이터 품질 확보 문제는 머신러닝 프로젝트에서 중요한 이슈이다
- 데이터 자체가 가지고 있는 의미는 없다
- 데이터를 측정할 때는 측정 도구가 신뢰할 만한 것인지를 확인해야한다

[ 데이터의 특징 ]
- Facts
- No meaning
- Representation of real world

[ Feature ]
feature는 데이터를 컴퓨터가 이해할 수 있도록 수치, 디지털로 표현 / 표상한 것이다.

* 동의어 : independent Variable, Explanatory Variable, Predictor, Input, Attribute

[ Model ]
실제의 무엇을 더 작게 추상화된 형태로 표현한 것, 모형 또는 본보기

[ Machine Learning Model ]
어떠한 문제를 해결하기 위해 수립한 가설을 논리적, 수학적 함수식의 형태로 표현한 것

[ Algorithm ]
입력된 자료를 바탕으로 원하는 결과를 유도하기 위해 일련의 논리적인 순서와 절차를 규칙화한 것

[ Machine Learning Algorithm ]
모델이 어떠한 문제를 해결하기 위한 함수식이라면 알고리즘은 그 함수식을 만들어내는 일련의 절차, 규칙

[ Learning ]
예측 목표로부터 예측 결과의 오차를 최소화하는 함수식을 찾아내는 과정

[ Target ]
머신러닝에서 예측하려는 목표, 도메인 전문가들과 함께 라벨링 작업을 진행한다

## 4_2 머신러닝을 위한 데이터 준비



데이터 preprocessing, feature engineering을 통해 좋은 모델을 만든다

## 4_3 머신러닝 프로젝트 성공의 열쇠

대다수의 data preprocessing과 feature engineering기법은 도메인에 많은 영향을 받는다

[ data preprocessing ]
- vectorization
- normalization
- handling missing values

[ feature engineering ]
도메인 지식을 활용하여 머신러닝 알고리즘이 학습을 잘 진행할 수 있도록 preprocessed data를 변환하는 작업

[ types of feature ]

* numerical : age, height, price
* categorical : gender, class, job

# 5. 머신러닝 실무 체험

## 5_1 문제 정의

[ 머신러닝의 종류 ]
* 지도학습 : input, output
* 비지도학습 : only input ( group, interpret data )




* 1_supervised Learning

1. classification : 분류, 범주 예측

2. Regression : 회귀, 숫자를 예측

* 2_unsupervised Learning
clustering

기타 머신러닝 문제


- 추천 시스템
- 이상탐지
- 네트워크 분석
- 차원 축소
- 프로파일링
- 시계열 에측 분석

## 5_2 효과 검증

효과 검증 설계
: 문제정의 -> 가설 설정 -> 해결 방안 -> 효과 검증




[ 공장에서 ]
* 문제 정의 : 공장에서 고장을 줄이기 위해서( 비용 감소, 수익 증가 )
* 가설설정 : 센서가 주기적으로 변하므로 이를 확인하고, 사전 진단을 통해 점검을 할 수 있다 -> 이상탐지 예측 모델 만든다 
* 해결 방안 : 예측 확률값 사용
* 효과 검증 : 실제로 고장이 나는지 안 나는지를 확인하고, KPI의 증가율 확인
