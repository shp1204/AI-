# 7. 최댓값 직접 구하기

```python
# 리스트 nums를 넣었을 때, 최댓값을 반환(return)하는 함수 our_max를 작성해봅시다.
def our_max(my_list) :
    return sorted(my_list)[-1]


print(our_max([1, 2, 10, 9, 3, 7, 0, 99, 27, 85]))
```

# 8. 환상의 나라 네버랜드로

```python
# 대기시간이 담긴 리스트가 인자로 주어지면 조건을 만족하도록
# 타야하는 대기시간의 순서가 담긴 리스트를 반환하는 함수 neverland()를 작성해봅시다.
def neverland(my_queue) :
    new_queue = [my_queue.pop(2)]
    new_queue[1:] = sorted(my_queue)
    return new_queue



# 확인을 위한 코드입니다.
# 대기시간이 담긴 리스트 queue를 자유롭게 수정해보세요!
queue = [30, 10, 20, 50, 40, 60]
print(neverland(queue))
```

# 9. 비밀번호 486

```python
# 비밀번호를 만들어주는 함수 yoonHa()를 만들어봅시다.
def yoonHa(nums) :

    secret = {'4':'love', '8':'smile', '6':'kiss'}
    
    result = ''
    for num in list(str(nums)):
        result += secret[num]
        
    return result


# 채점을 위한 코드입니다. 이를 수정하지 마세요!
nums = int(input())
print(yoonHa(nums))
```
