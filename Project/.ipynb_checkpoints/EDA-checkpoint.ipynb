```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd
import numpy as np
from datetime import datetime
import time

# 한글 폰트 설정

fm.get_fontconfig_fonts()
font_location = 'C:/Windows/Fonts/NanumGothic.ttf' # For Windows
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)
```

```python
# 데이터 로드
data = pd.read_csv('data/train.csv', encoding='cp949', header=0, parse_dates=["방송일시"])

# 상품군 = 무형 제거 ( 931 )

data.drop(data[data['상품군']=='무형'].index, inplace=True)
data.reset_index(inplace=True)


# 판매단가, 취급액 쉼표 제거 , int 변환

data['판매단가'] = data['판매단가'].str.replace(',','').astype(int)
data[' 취급액 '] = data[' 취급액 '].str.replace(',','').astype(int)
data['노출(분)'] = data['노출(분)'].astype(int)


# 방송일시 column 분리 ( , 일, 시, 분, 요일 )

data['방송일시'] = pd.to_datetime(data['방송일시'], format='%Y-%m-%d %H:%M:%S', errors='raise')

data['month'] = data['방송일시'].dt.month
data['day'] = data['방송일시'].dt.day
data['hour'] = data['방송일시'].dt.hour
data['minute'] = data['방송일시'].dt.minute
data['요일'] = data['방송일시'].dt.weekday_name 

# 06 ~ 익일 02 가 한 cycle임을 고려하여 0, 1, 2 에 +24를 해주자
def plus_24(x) :
    if x in [0, 1, 2] :
        return x + 24
    else :
        return x
    
data['hour'] = list(map(plus_24, data['hour']))    
```

```python
# use data
df = data.loc[:,['상품명','상품군','판매단가','month','day','hour','minute','요일','노출(분)',' 취급액 ']]
df
```

# 취급액

## 상품군별 취급액

* 농수축, 이미용 상품군의 취급액이  가장 높음

* 침구, 가구 상품군의 취급액이 가장 낮음

가장 편성이 많이 이루어지는 곳이 주방, 가전인데 비해 취급액의 평균은 높지 않은 편이다.

=> 가장 많이 팔리는 시간대만 공략하고 그렇지 않은 곳은 다른 제품으로 대체할 필요가 있지 않을까 ? 

```python
plt.figure(figsize=(20,5))
df[' 취급액 '].groupby(df['상품군']).mean().plot(kind='bar')
```

```python
plt.figure(figsize=(20,5))
df[' 취급액 '].groupby(df['상품군']).count().plot(kind='bar')
```

## 월별 취급액

* 4월 최저, 11월과 12월 최고의 취급액을 보인다

4월은 날씨가 좋아서 외출이 많음

11, 12월은 선물 및 집에 있는 시간이 많아지면서 ?

```python
plt.figure(figsize=(20,5))
df[' 취급액 '].groupby(df['month']).mean().plot(kind='bar')
```

```python
plt.figure(figsize=(20,5))
df[' 취급액 '].groupby(df['month']).count().plot(kind='bar')
```

## 시간대별

저녁시간대 count의 개수가 많음 => 편성 개수가 많음 => 짧게 편성되는 프로그램이 많다 => 비싼 가격 x ( 이시간대 주요 tv 사용 연령 고려할 필요가 있다 )

오후시간대는 count의 개수가 적은 반면 취급액의 크기는 큼 => 주부들을 고려한 편성

```python
plt.figure(figsize=(20,5))
df[' 취급액 '].groupby(df['hour']).mean().plot(kind='bar')
```

```python
plt.figure(figsize=(20,5))
df[' 취급액 '].groupby(df['hour']).count().plot(kind='bar')
```

## 요일별 취급액
편성 개수는 비슷하지만, 주말 취급액이 큼 => 고가제품 많이 편성

```python
plt.figure(figsize=(20,5))
df[' 취급액 '].groupby(df['요일']).count().plot(kind='bar')
```

```python
plt.figure(figsize=(20,5))
df[' 취급액 '].groupby(df['요일']).mean().plot(kind='bar')
```

# 상품군별

```python
clrs = ['c','m','y','b','lime','violet','g']

def cnt_product_day(day, color) :
    plt.figure(figsize=(20,5))
    plt.rc('font', size=20)
    df[df['요일']==day].groupby(df['상품군']).count()[' 취급액 '].plot(kind='bar', color=color)
    plt.title(day)
    plt.show()
    
for day, color in zip(df['요일'].unique(), clrs) :
    cnt_product_day(day, color)
```

```python
def mean_product_day(day, color) :
    plt.figure(figsize=(20,5))
    plt.rc('font', size=20)
    df[df['요일']==day].groupby(df['상품군']).mean()[' 취급액 '].plot(kind='bar', color=color)
    plt.title(day)
    plt.show()
    
for day, color in zip(df['요일'].unique(), clrs) :
    mean_product_day(day, color)
```

```python
plt.figure(figsize=(20,10))
plt.rc('font', size=20)
for day, color in zip(df['요일'].unique(), clrs) :
    df[df['요일']==day].groupby(df['상품군']).count()[' 취급액 '].plot(color=color)
```
