### 1. 필요한 package 준비

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# 그래프 그리기 위한 pakage
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# plot의 글자 크기, 제목 크기, 그래프 크기 설정
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.titlesize'] = 16
matplotlib.rcParams['figure.figsize'] = [9, 7]
```

### 2. 데이터 생성

```python
np.random.seed(2020) # random seed 28, 33, 52

# True function 선언
def true_gen(x):
    y = np.sin(1.2 * x * np.pi)
    return(y)

# random noise가 들어간 데이터 생성
x = np.sort(np.random.rand(120)) # 0~1 균일분포
y = true_gen(x) + 0.1 * np.random.randn(len(x)) # 표준정규분포
```

```python
# 만든 데이터를 training과 test 데이터로 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True, random_state=0)

# 이제까지 만든 true function과 데이터 가지고 그리기
x_linspace = np.linspace(0, 1, 100)
y_true = true_gen(x_linspace)

plt.plot(x_train, y_train, 'ko', label = 'Train')
plt.plot(x_test, y_test, 'ro', label = 'Test')
plt.plot(x_linspace, y_true, 'b-', linewidth = 2, label = 'True Function')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data')
```

### 3. Polynomial 모델

주어진 데이터에 대한 다항함수를 생성하는 모델입니다.

```python
def fit_poly(x_train, y_train, x_test, y_test, degrees, plot=False, return_errors=False):
    
    # 주어지는 degree에 따른 다항함수 생성
    features = PolynomialFeatures(degree=degrees, include_bias=False)
    
    # Training 데이터를 reshape해서 features로 fit & transform
    x_train = x_train.reshape((-1, 1))
    train_trans = features.fit_transform(x_train)

    # 선형 회귀를 이용해서 학습
    model = LinearRegression()
    model.fit(train_trans, y_train)

    # Training 데이터 prediction/error
    train_predictions = model.predict(train_trans)
    training_error = mean_squared_error(y_train, train_predictions)
  
    # Test 데이터를 reshape해서 features로 fit & transform
    x_test = x_test.reshape((-1, 1))
    test_trans = features.fit_transform(x_test)
    
    # Test 데이터 prediction/error
    test_predictions = model.predict(test_trans)
    test_error = mean_squared_error(y_test, test_predictions)

    # 모델링한 함수와 True function 곡선
    x_curve = np.linspace(0, 1, 100)
    x_curve = x_curve.reshape((-1, 1))
    x_curve_trans = features.fit_transform(x_curve)

    model_curve = model.predict(x_curve_trans)
    y_true_curve = true_gen(x_curve[:, 0])
    
    # train 데이터와 prediction, 그리고 true function 그리기
    if plot == 'train':
        plt.plot(x_train[:, 0], y_train, 'ko', label = 'Train Observations')
        plt.plot(x_curve[:, 0], y_true_curve, linewidth = 4, label = 'True Function')
        plt.plot(x_curve[:, 0], model_curve, linewidth = 4, label = 'Model Function')
        plt.ylim(-1, 1.5)
        plt.xlim(0, 1)
        plt.legend()
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('{} Degree Model on Training Data'.format(degrees))
        plt.show()
        
    # test 데이터와 prediction 그리기
    elif plot == 'test':
        plt.plot(x_test, y_test, 'o', label = 'Test Observations')
        plt.plot(x_curve[:, 0], y_true_curve, 'b-', linewidth = 2, label = 'True Function')
        plt.plot(x_test, test_predictions, 'ro', label = 'Test Predictions')
        plt.ylim(-1, 1.5)
        plt.xlim(0, 1)
        plt.legend()
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('{} Degree Model on Test Data'.format(degrees))
        plt.show()

    # Return training/test error
    if return_errors:
        return training_error, test_error
```

### 4. 모델의 degree = 1

```python
fit_poly(x_train, y_train, x_test, y_test, degrees=1, plot='train')
```

```python
fit_poly(x_train, y_train, x_test, y_test, degrees=1, plot='test')
```

```python
# 모델의 training error, testing error을 출력합니다.
errors = fit_poly(x_train, y_train, x_test, y_test, degrees=25, return_errors=True)
print("Training error: {:.4f}".format(errors[0]))
print("Test error: {:.4f}".format(errors[1]))
```

---

## 정리: 모델의 차수에 따른 Train, Test error

```python
# 차수 degree 1 ~ 40
degrees = [x for x in range(1, 41)]

# error를 저장할 dataframe 생성
results = pd.DataFrame(0, columns = ['train_error', 'test_error'], index = degrees)
# print(results.head())
for degree in degrees:
    degree_results = fit_poly(x_train, y_train, x_test, y_test, degree, return_errors=True)
    results.loc[degree, 'train_error'] = degree_results[0]
    results.loc[degree, 'test_error'] = degree_results[1]
print(results.head())
```

```python
# degree에 따른 training / test error 그리기
plt.plot(results.index, results['train_error'], 'b-o', ms=6, label = 'Training Error')
plt.plot(results.index, results['test_error'], 'r-*', ms=6, label = 'Test Error')
plt.legend(loc=1)
plt.xlabel('Degrees')
plt.ylabel('Mean Squared Error')
plt.title('Training and Test Curves');
plt.ylim(0, 0.05)
plt.show()

print('\nMinimum Training Error occurs at {} degrees.'.format(int(np.argmin(results['train_error']))))
print('Minimum Test Error occurs at {} degrees.\n'.format(int(np.argmin(results['test_error']))))
```
