## 3. ë‹¨ìˆœ ì„ í˜• íšŒê·€ í‰ê°€

### 3-1. RMSE, RMAE

ëª¨ë¸ ìµœì í™”ê°€ ë˜ì—ˆë‹¤ë©´ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¥¼ ë°”íƒ•ìœ¼ë¡œ ëª¨ë¸ì„ í‰ê°€í•©ë‹ˆë‹¤. 

#### RMSE

Root Mean Square Errorì˜ ì•½ìë¡œ ì¶”ì • ê°’ ë˜ëŠ” ëª¨ë¸ì´ ì˜ˆì¸¡í•œ ê°’ê³¼ ì‹¤ì œ í™˜ê²½ì—ì„œ ê´€ì°°ë˜ëŠ” ê°’ì˜ ì°¨ì´ë¥¼ ë‹¤ë£° ë•Œ í”íˆ ì‚¬ìš©í•˜ëŠ” ì²™ë„ì…ë‹ˆë‹¤.

##### RMSE ì •ì˜

> $$RMSE(Y,\widehat{Y})=\sqrt{E[(Y-\widehat{Y})^{2}]}$$,

> $$Y=[y_1, y_2,..., y_N], \widehat{Y}=[\widehat{y_1}, \widehat{y_2},..., \widehat{y_N}]$$,

$N$ê°œì˜ ìƒ˜í”Œì´ ìˆëŠ” ê²½ìš° $y_i$ëŠ” $i$ë²ˆì§¸ ìƒ˜í”Œì˜ ë ˆì´ë¸”(label) ê°’, $\widehat{y_i}$ëŠ” $i$ë²ˆì§¸ ìƒ˜í”Œì˜ ì˜ˆì¸¡ê°’ì„ ì˜ë¯¸í•©ë‹ˆë‹¤.

RMSE ê°’ì€ ë ˆì´ë¸” ê°’ì— ì–¼ë§ˆë‚˜ ì˜ ì˜ˆì¸¡í•˜ëƒì— ë”°ë¼ 0ì— ìˆ˜ë ´í•œ ê°’ì„ ê°–ê²Œë©ë‹ˆë‹¤.

#### RMAE

Root Mean Absolute Errorì˜ ì•½ìë¡œ RMSEì™€ ë¹„ìŠ·í•œ ë°©ì‹ì´ì§€ë§Œ errorì˜ square ê°’ ëŒ€ì‹  absolute ê°’ì„ ë‹¤ë£¬ë‹¤ëŠ” ì ì—ì„œ ì°¨ì´ê°€ ìˆìŠµë‹ˆë‹¤. 

##### RMAE ì •ì˜

> $$RMAE(Y,\widehat{Y})=\sqrt{E[\| Y-\widehat{Y}\|]}$$.

RMAE ê°’ ë˜í•œ RMSEì™€ ë§ˆì°¬ê°€ì§€ë¡œ ë ˆì´ë¸” ê°’ì— ì–¼ë§ˆë‚˜ ì˜ ì˜ˆì¸¡í•˜ëƒì— ë”°ë¼ 0ì— ìˆ˜ë ´í•œ ê°’ì„ ê°–ê²Œë©ë‹ˆë‹¤.

##### <ì˜ˆì œ 1>  RMSE, RMAE ê°’ ë¹„êµ

`4-2. ë‹¨ìˆœ ì„ í˜• íšŒê·€ class`ì—ì„œ `<ì˜ˆì œ 2>`ì˜ ê²°ê³¼ ì½”ë“œë¥¼ ê°€ì ¸ì™€ í•™ìŠµëœ ëª¨ë¸ì˜ RMSE ê°’ê³¼ RMAE ê°’ì„ ë¹„êµí•´ ë´…ì‹œë‹¤.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

feature_data = np.array([1,2,3,4]).reshape((-1,1))
label_data = np.array([3.1,4.9,7.2,8.9]).reshape((-1,1))

# ëª¨ë¸ ì„¤ì •
sci_model = LinearRegression()

# í•™ìŠµ ìˆ˜í–‰
sci_model.fit(feature_data, label_data)

# scikit-learn ì—ì„œëŠ” loss í•¨ìˆ˜ê°€ ëª¨ë¸ì•ˆì— ë‚´ì •ë˜ì–´ ìˆì§€ ì•Šê¸°ì— ì •ì˜
def loss(prediction, label):
        
    error = label - prediction
    ls = np.mean(error**2)

    return ls

print("loss: {}\n".format(loss(sci_model.predict(feature_data), label_data))) 

def RMSE(prediction, label):
    
    error = label - prediction
    ls = np.sqrt(np.mean(error**2))
        
    return ls

def RMAE(prediction, label):
    
    error = label - prediction
    ls = np.sqrt(np.mean(np.abs(error)))
        
    return ls

print("RMSE: {}\n".format(RMSE(sci_model.predict(feature_data), label_data))) 
print("RMAE: {}\n".format(RMAE(sci_model.predict(feature_data), label_data))) 
```

### 3-2. R2, adjusted R2

#### R2

$R^2$ëŠ” íšŒê·€ ëª¨ë¸ í‰ê°€ë¥¼ í•  ë•Œ ë§¤ìš° ìì£¼ ì“°ì´ëŠ” ë°©ë²•ì…ë‹ˆë‹¤.

$R^2$ ì•„ë˜ì™€ ê°™ì€ ìˆ˜ì‹ìœ¼ë¡œ ê³„ì‚°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

##### R2 ì •ì˜

> $$R^{2} = 1-\frac{\sum_{i=1}^{N}(y_i-\widehat{y_i})^{2}}{\sum_{i=1}^{N}(y_i-\bar{y_i})^{2}}$$,

> $$\bar{y_i}=\frac{1}{N}\sum_{i=1}^{N}y_i$$.

$R^2$ëŠ” errorì˜ ì œê³±ì„ ì‚¬ìš©í•œë‹¤ëŠ” ì ì—ì„œëŠ” MSEì™€ ë¹„ìŠ·í•˜ì§€ë§Œ ë¶„ëª¨ë¥¼ í‰ê· ê³¼ì˜ ì°¨ì˜ ì œê³±ìœ¼ë¡œ ë‚˜ëˆˆ ê²ƒì´ ë‹¤ë¦…ë‹ˆë‹¤.

ê·¸ë ‡ë‹¤ë©´ ë¶„ëª¨ì˜ ê°’ì€ ì–´ë–¤ ì˜ë¯¸ë¥¼ í• ê¹Œìš”?

label ê°’ê³¼ ê·¸ í‰ê· ì˜ ì°¨ì˜ ì œê³±ì˜ í‰ê· ì€ ë°”ë¡œ ë¶„ì‚°ê³¼ ê°™ì€ ì˜ë¯¸ë¼ê³  ìƒê°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ë”°ë¼ì„œ label ë°ì´í„°ê°€ ë§ì´ í©ì–´ì ¸ ìˆì„ ìˆ˜ë¡ ê·¸ ê°’ì´ ì»¤ì§€ê²Œ ë©ë‹ˆë‹¤.

ê²°ê³¼ì ìœ¼ë¡œ ë§ì´ í©ì–´ì ¸ ìˆëŠ” ì •ë„ì— ë”°ë¼ label ê°’ê³¼ ì˜ˆì¸¡ ê°’ì˜ ì˜í–¥ë ¥ì„ ë¯¸ì¹˜ê²Œ ë©ë‹ˆë‹¤.

ì •ë¦¬ë¥¼ í•˜ìë©´ ëª¨ë¸ì´ label ê°’ê³¼ ë¹„ìŠ·í•  ìˆ˜ë¡ label ë°ì´í„°ì˜ ë¶„ì‚°ì´ ë†’ì„ ìˆ˜ë¡ 1ì— ê°€ê¹Œìš´ ê°’ì„ ê°€ì§€ê²Œ ë©ë‹ˆë‹¤.

#### Adjusted R2

$R^2$ì— ëª¨ë¸ ë³µì¡ë„ì— ëŒ€í•œ íŒ©í„°ë¥¼ ì¶”ê°€í•œ ê²ƒì´ $Adjusted R^2$ ì…ë‹ˆë‹¤.

##### Adjusted R2 ì •ì˜

> $$Adjusted R^2 = 1-(1-R^2)\frac{N-1}{N-p-1}$$,

$N$ì€ ìƒ˜í”Œì˜ ê°œìˆ˜, $p$ëŠ” feature columnì˜ ê°œìˆ˜ë¥¼ ì˜ë¯¸í•©ë‹ˆë‹¤.

ë”°ë¼ì„œ ìƒ˜í”Œì´ ì‘ì„ìˆ˜ë¡, feature columnì´ ë§ì„ ìˆ˜ë¡ ë³´ë‹¤ ëª¨ë¸ì˜ ë³µì¡ì„±ì´ ë†’ì•„ì§€ê³  ë¶€ì •í™•í•¨ì„ ìœ„ ìˆ˜ì‹ ì²˜ëŸ¼ í‘œí˜„í–ˆìŠµë‹ˆë‹¤.

##### <ì˜ˆì œ 2> ğ‘…2 , ğ´ğ‘‘ğ‘—ğ‘¢ğ‘ ğ‘¡ğ‘’ğ‘‘ ğ‘…2 ê°’ ë¹„êµ

`4-2. ë‹¨ìˆœ ì„ í˜• íšŒê·€`ì—ì„œ `<ì˜ˆì œ 2>`ì˜ ê²°ê³¼ ì½”ë“œë¥¼ ê°€ì ¸ì™€ í•™ìŠµëœ ëª¨ë¸ì˜ $R^2$, $Adjusted R^2$ë¥¼ ê³„ì‚°í•˜ì—¬ ë´…ì‹œë‹¤.

```python
def R2(prediction, label):
    error = label - prediction
    r2 = 1-(np.sum(error**2)/np.sum((label-np.mean(label))**2))
    return r2

def AR2(prediction, feature, label):
    error = label - prediction
    r2 = 1-(np.sum(error**2)/np.sum((label-np.mean(label))**2))
    ar2 = 1-(1-r2)*(label.size-1)/(label.size-feature.shape[1]-1)
    return ar2

print("R2: {}\n".format(R2(sci_model.predict(feature_data), label_data))) 
print("Adjusted R2: {}\n".format(AR2(sci_model.predict(feature_data), feature_data, label_data))) 
```
