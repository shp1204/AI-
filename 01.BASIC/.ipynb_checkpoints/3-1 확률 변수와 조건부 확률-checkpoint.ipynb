## 1. 확률 변수와 조건부 확률

머신러닝 기법의 대부분은 통계학으로 이루어져 있다 해도 과언이 아닐 정도로 머신러닝과 통계학은 연관이 깊습니다.

데이터 자체의 통계적 특성을 찾았다면 비슷한 데이터를 추정할 수 있듯이, 머신러닝이라는 과정 자체를 하나의 통계적 분석으로 볼 수 있습니다. 

이번 장에서는 머신러닝에서 주로 사용되는 확률, 통계의 개념을 알아보겠습니다.

### 1-1. 확률과 확률 변수

보드 게임이라는 예시를 통해 확률과 확률 변수의 정의에 대해 알아보겠습니다.

보드 게임을 즐길 때면, 항상 다음 내가 뽑을 패나 던져서 나올 주사위 수에 대해서 고려하곤 합니다.

어떤 결과가 나올지에 대해서 조금이라도 예상할 수 있다면, 게임을 이기기 쉽기에 옛날 사람들은 그 방법을 고민했습니다.

이러한 물음에 답하기 위해서 나온 개념이 확률입니다. 

확률은 다음에 내가 뽑을 패나 던질 주사위에 대해서 어떤 결과가 나올지 수치로 비교를 할 수 있게 해주는 도구입니다.

그렇다면 확률은 어떻게 구할 수 있을까요?

확률의 정의는 다음과 같습니다.

##### 확률의 정의

> $$어떤\;사건의\;확률 = (그\;사건이\;일어날\;수\;있는\;경우의\;수) / (총\;경우의\;수)$$

보드게임에서의 주사위를 예를 들어봅시다. 주사위 숫자 6이 나올 확률을 구하기 위해서는 먼저 모든 경우의 수 부터 구해야 합니다.

주사위를 던졌을 때 나오는 면의 숫자는 1,2,3,4,5,6 이며 총 6가지의 경우가 있습니다.

그 경우 중, 주사위 숫자 6이 나올 경우의 수는 단 한 가지기에 확률의 값은 $\frac{1}{6}$으로 구할 수 있습니다.

한 가지 예를 더 들어봅시다. 아래와 같이 트럼프 카드가 있을 경우 A 카드를 뽑을 확률은 얼마나 될까요?

뽑을 수 있는 모든 카드의 수는 총 경우의 수로, 총 $(10+3)*4 = 52$ 가지의 수가 있습니다.

A 카드에 대한 경우의 수는 각 모양별 A가 뽑히는 경우이므로 총 4개 입니다.

따라서 A가 뽑힐 확률은 $\frac{4}{52}=\frac{1}{13}$입니다. 

트럼프 카드 중에서 A가 뽑힐 확률 이라고 매번 쓰는 것이 불편할 수 있기에 우리는 확률변수라는 것을 정의해서 확률에 대한 표현을 보다 명확하게 합니다.

확률변수는 특정 사건들에 대해서 실수로 대응 시킨 것을 의미합니다.

위 트럼프 카드에서는 A가 뽑히는 경우를 1, 숫자 2가 뽑힐 경우를 2,... king이 뽑힐 경우를 13으로 확률변수를 정의한다면, 확률변수는 $X\in \{1,2,3,4,..,13\}$으로 표현 가능합니다.

따라서 확률변수 $X=1$의 의미는 A가 뽑히는 사건이랑 동일한 의미를 갖습니다.

자 이제 확률변수를 활용하여 A가 뽑힐 확률을 표현하면 $P(X=1)$으로 나타낼 수 있습니다.

따라서, $P(X=13)$은 king이 뽑힐 확률이라 할 수 있습니다.

그렇다면, 클로버 A 카드가 뽑힐 확률은? 이라고 물어보면 어떻게 표현해야 할까요?

앞서 정의한 확률변수의 사건에는 이것이 표현되어 있지 않았습니다.

이런 경우에는 확률변수에 대해서 새롭게 정의해야 합니다.

클로버 A가 뽑힐 경우를 1, 클로버 2는 2,... 클로버 킹은 13, 아래 그림과 같이 정리할 수 있습니다.

그렇다면 $P(X=1)$의 의미는 클로버 A가 뽑힐 확률이 됩니다.

이 확률변수를 사용해서 A가 뽑힐 확률을 표현하면 어떻게 될까요?

사건 1,14,27,40이 나올 확률이기에 $P(X=1\;or\;14\;or\;27\;or\;40)$으로 표현할 수 있습니다.

확률과 확률 변수에 대한 개념을 이해했으니 코드로는 어떻게 표현 가능한지 아래 예를 수행하며 알아봅시다.

```python
# 사건 A와 사건 B 정의
A = [1, 2, 3, 4]
B = [2, 4, 6, 8, 10, 12]

# 전체 경우의 수 U
U = set(A+B)

# 확률 계산
prob_A = len(A)/len(U)
prob_B = len(B)/len(U)

print("probability of A: {}".format(prob_A))
print("probability of B: {}".format(prob_B))
```

#### 이산 변수와 연속 변수

확률 변수는 셀수 있는 숫자인가 아닌가로 나누어집니다.

위의 카드 예시처럼 1~52의 셀수 있는 수로 이루어진 변수를 **이산 변수(discrete variable)**라 합니다. 

그렇다면 셀수 없는 단위인 사람의 키나 몸무게 등을 확률 변수로 만든 것을 **연속 변수(continuous)**라 합니다.

### 1-2. 확률 분포

확률변수에 따라 각 확률값을 대응시킨 분포를 확률 분포라 합니다.

위에서 예시로 사용하였던 트럼프 카드로 다시 설명해드리도록 하겠습니다.

클로버 A가 뽑히는 사건인 $X=1$에 대한 확률은 전체 카드 중에 1장만 존재하는 클로버 A이기에 1/52로 구할 수 있습니다.

같은 방식으로 모든 카드들에 대한 확률을 구하게 되면 아래의 그림처럼 1/52로 동일하게 표현할 수 있습니다.

위 그림처럼 각 확률 변수에 대한 확률들의 분포를 **확률 분포(probabiliy distribution)**라 합니다.

확률 분포는 성적 분포에서 쉽게 볼 수 있는 가우시안 확률 분포, 여러 주사위를 던져 특정한 숫자가 몇번 나올지에 대한 확률을 나타내는 이항 확률 분포 등, 실생활 및 자연현상에 쉽게 볼 수 있는 확률 분포들은 수학자들이 그 특성들을 잘 정리해 두었습니다.

이러한 대표적인 확률 분포는 확률 변수가 이산 변수인지 연속 변수인지에 따라 이산 확률 분포와 연속 확률 분포로 나누어집니다.

**이산 확률 분포**

확률 변수가 이산 변수로 되어 있다면 이는 이산 확률 분포라고 합니다.

대표적인 이산 확률 분포는 아래와 같습니다.

1. 베르누이 분포(bernoulli distribution): $P(X=0)=p, \;\;P(X=1)=1-p, \;\; 0\leq p \leq 1$
    + 동전을 던졌을 때 앞면 또는 뒷면이 나올 확률과 같이 어떠한 사건이 발생하거나 발생하지 않은 것에 대한 확률 분포.


2. 이항 분포(bimodal distribution): $P(X=x)=\binom{n}{p}p^{x}(1-p)^{n-x}$
    + 베르누이 분포를 여러번 했을 때 그 사건이 몇번 발생한 것에 대한 확률 분포


3. 이산 균등 분포(discrete uniform distribution): $P(X=x)=\frac{1}{b-a+1}$
    + 모든 확률 변수에 대해서 똑같은 확률 값을 갖는 확률 분포

앞서서 확률 변수에 따른 확률 값을 확률 분포로 구하였다면, 반대로 확률 분포를 가지고 확률 변수의 샘플 데이터를 생성할 수 있습니다.

예를 들어 동전의 앞면이 나올 확률에 대해서 $\frac{1}{2}$ 값을 갖는 베르누이 분포가 있다면, 컴퓨터에게 실제로 이 확률 값을 바탕으로 동전 던지기를 수행할 수 있습니다.

##### <예제 1> 이산 확률 분포 샘플링

위 예와 같이 확률 분포를 바탕으로 샘플 데이터를 생성하는 것을 코딩으로 구현해봅시다.

```python
import numpy as np

# 베르누이 분포
# 베르누이는 시행 1번, p=0.5 확률
n, p = 1, .5

#  0.5 확률로 뽑힌 개수를 20번 수행하여 출력
bern_sample = np.random.binomial(n, p, 20)
print("베르누이 분포 샘플: {}".format(bern_sample))

# 이항 분포
# 10번 시행, p=0.3 확률
n, p = 10, .3

# 10번의 시행에서 0.3 확률로 뽑힌 개수를 20번 수행하여 출력
bimodal_sample = np.random.binomial(n, p, 20)
print("이항 분포 샘플: {}".format(bimodal_sample))

# 이산 균등 분포
# X=0,1,2.,...,10
uniform_sample = np.random.randint(0,10,20)
print("이산 균등 분포 샘플: {}".format(uniform_sample))
```

**연속 확률 분포**

확률변수가 연속 변수로 되어 있다면 이는 연속 확률 분포라고 합니다.

대표적인 연속 확률 분포는 다음과 같습니다.

1. 연속 균등 분포(continuous uniform distribution): $P(X=x)=\frac{1}{b-a}$
    + 연속 변수 $X$가 a보다 크고 b보다 작은 범위에서 정의된다면 모두 같은 확률 값을 갖는 확률 분포
    
    
2. 지수 분포(exponential distribution): $P(X=x)=\lambda e^{-\lambda x}$
    + 연속 변수의 크기의 제곱수만큼 증가하는 확률 분포
    
    
3. 정규 분포(normal or gaussian distribution): $P(X=x)=\frac{1}{\sigma\sqrt{2\pi }}e^{\frac{(x-\mu)^{2}}{2}}$
    + 평균 값에서 가장 높은 확률 값을 가지며 그 지점을 중심으로 대칭을 이룬 확률 분포

##### <예제 2> 연속 확률 분포 샘플링

이산 확률 분포와 마찬가지로 연속 확률 분포도 샘플 데이터를 생성해 봅시다.

```python
import numpy as np

# 연속 균등 분포
# X=0~10
uniform_sample = np.random.uniform(0,10,20)
print("연속 균등 분포 샘플: \n{}\n".format(uniform_sample))

# 지수 분포
# beta = 1/lambda
beta = 1

# 지수 분포를 따르는 샘플링을 20번 수행하여 출력
exp_sample = np.random.exponential(beta, 20)
print("지수 분포 샘플: \n{}\n".format(exp_sample))

# 정규 분포
# 평균:1, 표준편차: 0.1 
mu, sigma = 0, 0.1

normal_sample = np.random.normal(mu,sigma,20)
print("정규 분포 샘플: \n{}\n".format(normal_sample))
```

### 1-3. 조건부 확률

조건부 확률은 두 가지 사건에 대한 관계를 확률에 적용하기 위해 사용합니다.

두 가지 사건에 대한 설명을 위해서 주말 외출에 대한 예를 들어보겠습니다.

주말에 외출을 한다 안한다에 대한 사건 $A\in\{0(외출 함), 1(외출 안함)\}$을 정의한다면 주말에 나갈 확률은 $P(A=0)$ 으로 표현할 수 있습니다.

만약 주말에 약속을 잡았다 안잡았다에 대한 사건 $B\in\{0(약속함), 1(약속안함)\}$을 정의한다면 주말에 약속을 잡았을 때, 외출을 할 확률은 $P(A=0|B=0)$으로 표현합니다.

$P(A=0)$과 $P(A=0|B=0)$은 어떤 차이가 있을까요? 

일반적으로 주말에 약속을 잡은 상황에서 외출 할 확률이 약속이 잡힌지 안잡히지도 모르는 상황에서 외출을 할 확률보다 높을 겁니다.

위 예처럼 두 사건 간의 전 후 관계에 따라 확률은 달라지기에 이를 표현하기 위해서 조건부 확률을 정의합니다.

##### 조건부 확률 정의

> $$P(A=0|B=0)=\frac{P(A=0\;and\;B=0)}{P(B=0)}$$

양 변에 $P(B=0)$을 곱하면 다음과 같이 표현 가능합니다.

##### 조건부 확률의 다른 표현

> $$P(B=0)P(A=0|B=0)=P(A=0\;and\;B=0)$$

이는 **약속을 잡았을 확률**과 **약속을 잡았을때 외출을 할 확률**을 곱한 것은 **주말에 약속을 잡았고, 외출을 할 확률**을 의미합니다.

조건부 확률에 대해서 코딩을 통해서 알아봅시다.

##### <예제 3> 조건부 확률

사건 A와 사건 B에 대한 집합이 주어졌을 때, 각 사건들에 대한 조건부 확률을 구해봅시다.

```python
import numpy as np

# 사건 A와 사건 B 정의
A = [1, 2, 3, 4]
B = [2, 4, 6, 8, 10, 12]

# 전체 경우의 수 U
U = set(A+B)

# 확률 계산
prob_A = len(A)/len(U)
prob_B = len(B)/len(U)

# A,B 사건이 모두 해당될 확률
prob_A_and_B = len(np.intersect1d(A,B))/len(U)

# B 사건이 발생 했을 때, A 사건도 발생할 경우
conditional_P_A_B = prob_A_and_B/prob_B
# A 사건이 발생 했을 때, B 사건도 발생할 경우
conditional_P_B_A = prob_A_and_B/prob_A

print("P(A|B): {}".format(conditional_P_A_B))
print("P(B|A): {}".format(conditional_P_B_A)) 
```

`np.intersect1d(A,B)` 은 사건 A와 B의 교집합을 구하는 함수입니다. 

```python
import numpy as np

A = [1, 2, 3, 4]
B = [2, 4, 6, 8, 10, 12]
U = set(A+B)

len(B)/len(U)
```
