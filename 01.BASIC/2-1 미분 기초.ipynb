## 1. 미분법

미분의 기본적인 의미는 순간 변화량을 표현하는 것입니다. 

머신러닝에서 미분은 왜 필요할까요?

머신러닝에서 미분은 다양한 방식으로 활용되지만 가장 중요하게는 최대 값 또는 최소 값을 구하는 최적화 방식에서 사용합니다.

머신러닝 모델은 다양한 파라미터들을 사용하게 되는데 이 값을 변화시킴에 따라 예측 정확도가 바뀌게 됩니다.

이러한 파라미터를 조정하여 최대 정확도가 나오게 조절할 수만 있다면, 우리는 좋은 모델을 만들었다 생각할 수 있을 것입니다.

그렇기에 정확도를 최대로 만들거나 예측치와의 오차를 최소로 하는 문제를 풀어야 하고 거기서 미분법은 그 솔루션을 구하는데 중요한 도구가 될 것입니다.

### 1-1. 미분 정의

파라미터를 최적화와 미분과의 관계를 이해하기 앞서 미분의 정의부터 알아봅시다.

미분은 입력값의 순간 변화량에 따른 출력값의 변화량을 의미하며, 물리적으로는 얼마나 빠르게 변화하느냐의 대한 척도를 의미합니다.

예를 들어 성능 좋은 GPS가 여러분이 카페에서 집까지 가는 위치를 매시간 기록했다고 가정해보겠습니다.

GPS가 기록한 여러분의 위치 정보에 대해서 함수 $f$로 표현하면 아래와 같이 표현할 수 있습니다.

##### 위치 정보 함수 

> $$f(초 단위 시간)=위치 좌표$$

예를 들어 카페에서 막 출발하는 시간 0초라고 생각하면 $f(0)=카페위치좌표$ 라고 표현할 수 있습니다.

10분이 지나고 여러분은 집 근처인 편의점 앞을 지나고 있다고 가정해 봅시다.

그렇다면 10분동안 카페에서 편의점 앞까지의 거리를 움직였다고 생각할 수 있습니다. 

이걸 바탕으로 여러분은 다른사람에게 카페에서 집까지 갈때 10분동안 이동한 정도를 얘기해 줄 수 있습니다.

같은 방식으로 움직인 구간의 시간과 위치를 알게 되면 얼마나 빨리 이동하느냐에 대한 정보를 얻을 수 있습니다.

과연 여러분은 편의점까지 이동한 10분동안 계속 같은 속도로 움직였을까요? 중간에 횡단보도를 기다릴 수도 있고 중간 중간 뛸 수도 있습니다.

그렇기에 사람들은 궁금해진것입니다. 내가 5초, 10초, 3분 1초,...라는 **그 순간**에 얼마나 빨리 움직였을까 말이죠.

미분은 이 정보를 알려주는 도구입니다.

$f(시간)$을 시간에 대해서 미분한 함수를 도함수라 하고 $f'(시간)$ 라고 표현합니다.

그렇다면 $f'(5초)$의 출력 값은 5초에서 순간에서의 빠르기라고 할 수 있습니다.

이처럼 미분은 입력 값(시간)의 그 순간에 대한 출력 값(위치좌표)의 변화량(이동거리)을 표현하기 위해서 고안된 기술입니다.

이를 수학적으로 정리 하면 다음과 같이 표현 가능합니다.

##### 미분 정의

> $$\lim_{\Delta x \rightarrow 0}\frac{f(x + \Delta x) - f(x)}{\Delta x}$$

$x$는 입력 값을 의미하며, $\Delta x$는 매우 작은 입력 값의 변화량을 의미합니다.

위 수식의 아이디어는 아래의 그림으로 설명할 수 있습니다.

위 그림처럼 5초에서의 빠르기를 구하기 위해서 빠르기를 구하는 구간을 짧게 하여 빠르기를 구하고 있습니다.

만약 이 5초에서 그다음 측정 시간을 5.000001초보다 더 짧게 만든다면 조금 더 정확한 순간의 속도를 알 수 있을것입니다.

이를 일반화해서 x초에서의 빠르기를 표현한다면 아래와 같이 구할 수 있습니다.

##### 빠르기 함수

> $$빠르기(x)=\lim_{\Delta x \rightarrow 0}\frac{f(x + \Delta x) - f(x)}{x + \Delta x - x}$$

이 시간차의 구간의 크기를 $\Delta x $라 생각하게 되면, 결과적으로 0에 가까운 구간에서의 빠르기가 미분이란 것을 수식적 정의를 보며 이해할 수 있습니다.

### 1-2. 기초 미분법

위의 GPS 예를 조금 더 확장해서 생각해 봅시다.

만약 카페에서 움직이기 시작한 시점부터 점점 빠르게 달려서 집까지 달렸다고 합시다.

이 경우의 시간에 따른 위치 함수를 아래와 같이 정의했다고 가정합시다.

##### 시간에 따른 위치 함수

> $$f(시간(초))=시간^{2}(cm)$$

그렇다면 0초에는 0 cm, 10초에는 100 cm, 100초에는 10000 cm 위치에 있는 것을 알 수 있습니다.

이 경우 도함수를 구하려면 앞서 소개한 정의를 사용하여 $f'(시간(초))$를 구할 수 있습니다.

##### 도함수

> $$
\begin{aligned}
f'(시간(초))&=\lim_{\Delta 시간 \rightarrow 0}\frac{f(시간 + \Delta 시간) - f(시간)}{\Delta 시간}\;\; \\
&=\lim_{\Delta 시간 \rightarrow 0}\frac{(시간 + \Delta 시간)^{2} - 시간^{2}}{\Delta 시간}\;\;\;\; \\
&=\lim_{\Delta 시간 \rightarrow 0}\frac{\Delta 시간^{2} + 2시간\Delta 시간 }{\Delta 시간}\;\;\; \\
&=2시간
\end{aligned}$$

극한의 특성에 따라 $\lim_{\Delta 시간 -> 0}\Delta 시간 = 0$이기에 위와 같이 구할 수 있습니다.

하지만 이러한 과정을 매번 반복하기엔 계산량이 너무 많습니다.

이를 해결하기 위해서 아래와 같은 미분 공식을 사용합니다.

##### 미분 공식

> $$\begin{aligned}
f(x)=c &=> f'(x)=0 \;\;(c는 상수)\\
f(x)=x^{n} &=> f'(x)=nx^{n-1} \\ 
f(x)=e^{x} &=> f'(x)=e^{x} \\ 
f(x)=\sin(x) &=> f'(x)=\cos(x) \\
f(x)=\ln(x) &=> f'(x)=\frac{1}{x} \\
cf(x) &=> cf'(x) \\
f(x)+g(x) &=> f'(x)+g'(x) \\
f(x)g(x) &=> f'(x)g(x)+f(x)g'(x)
\end{aligned}$$

### 1-3. 미분 코딩하기

미분을 수행해야 할 함수만 알고 있을 경우, 코드 작성을 통해 미분 값 및 도함수를 구할 수 있는 방법을 소개합니다.

#### Sympy 사용

수학을 자주 다루지 않은 사람들은 미분을 공식을 이해하는 것조차 쉽지 않을 수 있습니다.

다행히 python library 인 sympy는 이 문제를 코드 몇줄로 간단하게 해결합니다.

##### <예제 1> sympy 사용

```python
import numpy as np
import sympy 

x = sympy.symbols('x') 

# f(x) 함수를 설정
sym_f = sympy.expand((x-1)**2)
print("Before differentiation : {}".format(sym_f))

# Sympy 계산 함수
sym_d_f_x = sympy.diff(sym_f,x)
print("Sympy differentiation by x : {}".format(sym_d_f_x)) 
```

sympy로 미분을 사용하기 위해서는 먼저 변수로 사용될 symbol을 설정해야 합니다.

위 예처럼 `sympy.symbols()` 함수를 사용하여 변수를 등록합니다.

이후 미분을 수행할 함수를 `sympy.expand()`를 사용하여 설정합니다.

이제 미분을 수행할 `sympy.diff()`에 미분할 함수 sym_f를 입력하여 결과를 출력하면 미분된 도함수의 형태를 알 수 있습니다.

여기서 중요한 점은 sym_d_f_x는 도함수의 형태가 어떤지만을 출력하지 실제로 값을 계산해주는 없다는 것입니다.

따라서 결과로 나온 `2*x - 2`을 다시 함수로 만들어야 도함수의 값을 계산할 수 있습니다.

```python
def d_f_x(x):
    return 2*x-2

print("f'(1): {}".format(d_f_x(1)))
```

#### $\Delta x$ 사용하기

이번엔 수학적 정의를 이용하여 도함수의 값을 구해보겠습니다.

##### <예제 2> 델타 x 사용하기

```python
# 미분을 수행할 함수
def f_x(x):

    return (x-1)**2

# delta x를 사용하여 미분
def d_f_x_incremental(x):

    x_incremental = 0.00001

    return (f_x(x + x_incremental) - f_x(x))/x_incremental


print("f'(1) = {}".format(d_f_x_incremental(1)))
```

참고로 e-05는 10의 -5승을 의미합니다.

실제 미분에서는 $\Delta x$의 값이 0으로 수렴하기에 100% 정확한 값을 출력할 순 없지만 `x_incremental`값을 작게 정의할수록 정확한 미분 값을 계산할 수 있습니다.
