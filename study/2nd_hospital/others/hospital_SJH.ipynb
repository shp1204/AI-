### Preprocessing

```python
# "OC"를 1과 0으로 변경. 1은 open, 0은 close
train.loc[train["OC"] == "open", "OC"] = 1
train.loc[train["OC"] == " close", "OC"] = 0
```

```python
test = test.drop("OC", axis=1)
```

```python
train.loc[train["ownerChange"] == "same", "ownerChange"] = 1
train.loc[train["ownerChange"] == "change", "ownerChange"] = 0

test.loc[test["ownerChange"] == "same", "ownerChange"] = 1
test.loc[test["ownerChange"] == "change", "ownerChange"] = 0
```

```python
test["employee1"] = test["employee1"].astype('str').str.replace(",","").astype("float")
test["employee2"] = test["employee2"].astype('str').str.replace(",","").astype("float")
```

```python
# 트레이닝 데이터를 open과 close로 나눔
train_open = train.loc[train["OC"] == 1]
train_close = train.loc[train["OC"] == 0]
```

```python
# 상대적으로 데이터가 많은 open에서만 결측치 있는 row를 제거
nan_index_list = train_open.loc[train_open["revenue1"].isnull()].index
train_open = train_open.drop(nan_index_list)

nan_index_list = train_open.loc[train_open["employee1"].isnull()].index
train_open = train_open.drop(nan_index_list)

nan_index_list = train_open.loc[train_open["employee2"].isnull()].index
train_open = train_open.drop(nan_index_list)

nan_index_list = train_open.loc[train_open["ownerChange"].isnull()].index
train_open = train_open.drop(nan_index_list)

nan_index_list = train_open.loc[train_open["bedCount"].isnull()].index
train_open = train_open.drop(nan_index_list)
```

```python
train = pd.concat([train_open, train_close])
```

### XGBoost

```python
# 트레이닝에 시용할 피쳐 리스트
label_list = ["bedCount","employee1","employee2","interest1","interest2","profit1","profit2","netAsset1","debt1"]

X_train = train[label_list]
X_test = test[label_list]
# X_train = train.drop(["OC","inst_id","sido","sgg","openDate","instkind","ownerChange"], axis=1)
# X_test = test.drop(["OC","sido","sgg","inst_id","openDate","instkind","ownerChange"], axis=1)
y_train = train["OC"]
```

```python
import xgboost as xgb
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
xgb_model = xgb.XGBClassifier(booster='gbtree',
                       scale_pos_weight=1,
                       learning_rate=0.05,
                       colsample_bytree = 0.4,
                       subsample = 0.8,
                       objective='binary:logistic',
                       n_estimators=500,
                       max_depth=5,
                       seed=777)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
xgb_pred = xgb_model.predict(X_test)
```

```python
hr_dmatrix = xgb.DMatrix(data=X_train, label=y_train)

cv_params = {"objective":"binary:logistic",
             "max_depth":4}

hr_cv = xgb.cv(dtrain=hr_dmatrix, 
               params=cv_params, 
               nfold=4, 
               num_boost_round=10, 
               metrics="auc", 
               as_pandas=True)

hr_cv
```
